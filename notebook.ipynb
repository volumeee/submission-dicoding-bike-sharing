## Data Wrangling

### Gathering Data

```python
df_hour = pd.read_csv('hour.csv')
df_day = pd.read_csv('day.csv')
```

### Assessing Data

```python
print(df_hour.info())
print(df_hour.describe())

print(df_day.info())
print(df_day.describe())
```

### Cleaning Data

```python
print(df_hour.isnull().sum())
print(df_day.isnull().sum())
```

## Exploratory Data Analysis (EDA)

## Explore weather situation

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='weathersit', y='cnt', data=df_hour)
plt.title('Jumlah Peminjaman Sepeda berdasarkan Kondisi Cuaca', fontsize=16)
plt.xlabel('Kondisi Cuaca', fontsize=12)
plt.ylabel('Jumlah Peminjaman', fontsize=12)
plt.xticks([0, 1, 2, 3], ['Clear', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow'], fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Explore working day vs non-working day

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='workingday', y='cnt', data=df_hour)
plt.title('Jumlah Peminjaman Sepeda: Hari Kerja vs Hari Libur', fontsize=16)
plt.xlabel('Hari Kerja (1) vs Hari Libur (0)', fontsize=12)
plt.ylabel('Jumlah Peminjaman', fontsize=12)
plt.xticks([0, 1], ['Hari Libur', 'Hari Kerja'], fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pengaruh cuaca terhadap jumlah peminjaman sepeda?

```python
plt.figure(figsize=(12,6))
sns.barplot(x='weathersit', y='cnt', data=df_hour)
plt.title('Rata-rata Jumlah Peminjaman Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-rata Jumlah Peminjaman')
plt.xticks([0, 1, 2, 3], ['Clear', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow'], fontsize=10)
plt.show()
```

### Pertanyaan 2: Apakah ada perbedaan signifikan antara jumlah peminjaman sepeda pada hari kerja dan hari libur?

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='workingday', y='cnt', data=df_hour)
plt.title('Distribusi Jumlah Peminjaman Sepeda: Hari Kerja vs Hari Libur')
plt.xlabel('Hari Kerja (1) vs Hari Libur (0)')
plt.ylabel('Jumlah Peminjaman')
plt.xticks([0, 1], ['Hari Libur', 'Hari Kerja'], fontsize=10)
plt.show()
```

  - Terdapat perbedaan dalam distribusi jumlah peminjaman sepeda antara hari kerja dan hari libur.
  - Hari kerja cenderung memiliki jumlah peminjaman yang lebih tinggi dan konsisten, sementara hari libur memiliki variasi yang lebih besar dalam jumlah peminjaman.

```python
# Mengambil kolom yang diperlukan dari df_day
df_day_selected = df_day[['dteday', 'season', 'yr', 'mnth', 'holiday', 'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']]

# Menggabungkan df_hour dengan df_day_selected berdasarkan kolom 'dteday'
main_data = pd.merge(df_hour, df_day_selected, on='dteday', suffixes=('_hour', '_day'))

# Menghapus kolom yang duplikat atau tidak diperlukan
columns_to_drop = ['casual_hour', 'registered_hour', 'cnt_hour', 'season_hour', 'yr_hour', 'mnth_hour', 'holiday_hour', 'workingday_hour', 'weathersit_hour', 'temp_hour', 'atemp_hour', 'hum_hour', 'windspeed_hour']
main_data = main_data.drop(columns=columns_to_drop)

# Mengubah nama kolom agar lebih jelas
main_data = main_data.rename(columns={
    'casual_day': 'casual',
    'registered_day': 'registered',
    'cnt_day': 'total_count',
    'season_day': 'season',
    'yr_day': 'year',
    'mnth_day': 'month',
    'holiday_day': 'holiday',
    'workingday_day': 'workingday',
    'weathersit_day': 'weather_situation',
    'temp_day': 'temperature',
    'atemp_day': 'feels_like_temperature',
    'hum_day': 'humidity',
    'windspeed_day': 'windspeed'
})

# Menyimpan DataFrame gabungan ke CSV
main_data.to_csv('main_data.csv', index=False)

print("Shape of main_data:", main_data.shape)
print("Columns in main_data:", main_data.columns.tolist())
```
