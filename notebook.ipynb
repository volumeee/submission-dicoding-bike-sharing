#Advanced Analysis

##RFM Analysis

```python
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])
df_hour['datetime'] = df_hour['dteday'] + pd.to_timedelta(df_hour['hr'], unit='h')

today = df_hour['datetime'].max() + pd.Timedelta(hours=1)
rfm = df_hour.groupby('hr').agg({
    'datetime': lambda x: (today - x.max()).total_seconds() / 3600,
    'instant': 'count',
    'cnt': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

def normalize(x):
    if x.max() == x.min():
        return x * 0
    return (x - x.min()) / (x.max() - x.min())

rfm_normalized = rfm[['Recency', 'Frequency', 'Monetary']].apply(normalize)

rfm_normalized['Recency'] = 1 - rfm_normalized['Recency']

rfm['RFM_Score'] = (rfm_normalized['Recency'] * 0.2 +
                    rfm_normalized['Frequency'] * 0.4 +
                    rfm_normalized['Monetary'] * 0.4)

rfm = rfm.reset_index()

chart = alt.Chart(rfm).mark_circle().encode(
    x='Recency',
    y='Frequency',
    size=alt.Size('Monetary', scale=alt.Scale(range=[100, 1000])),
    color=alt.Color('RFM_Score:Q', scale=alt.Scale(scheme='viridis')),
    tooltip=['hr', 'Recency', 'Frequency', 'Monetary', 'RFM_Score']
).properties(
    width=600,
    height=400,
    title='RFM Analysis by Hour'
).interactive()

text = alt.Chart(rfm).mark_text(dx=10, dy=-10).encode(
    x='Recency',
    y='Frequency',
    text='hr:N'
)

# Combine chart and text
final_chart = chart + text

# Display the chart
final_chart.display()

# Display statistics
print(rfm.describe())
print("\nRFM Score Range:", rfm['RFM_Score'].min(), "-", rfm['RFM_Score'].max())
```

### Conclution RFM Analysis
Berdasarkan analisis RFM:
- Recency: Nilai berkisar dari 1.00 hingga 24.00 jam.
- Frequency: Nilai berkisar dari 697 hingga 730 penyewaan.
- Monetary: Nilai berkisar dari 4428.00 hingga 336860.00 total penyewaan sepeda.
- Skor RFM tertinggi adalah 0.95, menunjukkan jam yang paling berharga.
- Skor RFM terendah adalah 0.03, yang mungkin memerlukan strategi perbaikan khusus.


Wawasan tambahan:
- Jam paling berharga adalah 17:00, dengan Skor RFM 0.95.
- Jam paling kurang berharga adalah 3:00, dengan Skor RFM 0.03.
- Jam-jam dengan skor RFM tinggi mungkin merupakan target yang baik untuk promosi atau peningkatan ketersediaan layanan.
- Jam-jam dengan skor RFM rendah mungkin memerlukan strategi untuk meningkatkan penyewaan sepeda, seperti diskon khusus di luar jam sibuk.
